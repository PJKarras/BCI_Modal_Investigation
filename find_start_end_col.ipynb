```python
"""
Adjust start_col and end_col here until you pull out the part of the image you would 
like to analyze, use these values in the analysis py notebook
"""
import pandas as pd
import os
import numpy as np
from PIL import Image
from matplotlib import cm
import matplotlib.pyplot as plt

# Prints out single image
maxh = 41.8977567043499
minh = 22.584069668121902

saveName = "2015_06_01_11_00_bci_t"

# Data for grabbing particular part of image (here by horizontal strips of which 
# are vertical in saved files)
#(top row) 240 - 0 (bottom row)

# EDIT THESE VALUES
#
start_col = 230     
end_col = 130               
#
# ----------------

cols = ['V' + str(x) for x in range(start_col, end_col,-1)]

pd_thermal_img = pd.read_csv(saveName + ".csv",usecols=cols)
np_thermal_img_old = np.asarray(pd_thermal_img)
myarray = ((np_thermal_img_old-minh)/(maxh - minh))
print(myarray)
print(len(myarray[0]))

img = np.rot90(myarray)
fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(111)
#ax.imshow(img, vmin=0, vmax=1, cmap="bwr")
ax.imshow(img, vmin=0, vmax=1, cmap="CMRmap")
fig.savefig(("test_col_range_" + saveName), bbox_inches = "tight")
```
